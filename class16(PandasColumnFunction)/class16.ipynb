## Apply custom function on multiple columns and return one column

```python
df.head()
```

```python
def fn_obtained(s1:int,s2:int,s3:int,s4:int,s5:int)->int:
    return s1+s2+s3+s4+s5

df["obt"] = df[["s1","s2","s3","s4","s5"]].apply(lambda x:fn_obtained(*x), axis=1)
df
```

```python
def fn_obtained(s1:int,s2:int,s3:int,s4:int,s5:int)->tuple[int,int,float,str]:
    total : int = 500
    ob : int  = s1+s2+s3+s4+s5
    per : float = ob / total * 100
    grade : str = " "
    if per >= 80:
        grade =  "A+"
    elif per >= 70:
        grade =  "A"
    elif per >= 60:
        grade =  "B"
    elif per >= 50:
        grade =  "C"
    elif per >= 40:
        grade =  "D"
    elif per >= 33:
        grade =  "E"
    else:
        grade =  "F"

    return total,ob,per,grade    
    

df[["Total",'obtained', 'percentage', "Grade"]] = df[["s1","s2","s3","s4","s5"]].apply(lambda x:fn_obtained(*x), axis=1,result_type='expand')
df
```

## Map function for remaks

```python
from typing import Dict

remarks : Dict[str,str] = {
    "A+" : "Excellent",
    "A" : "Very Good",
    "B" : "Good",
    "C" : "Fair",
    "D" : "Poor",
    "E" : "Very Poor",
    "F": "Fail"
}

df["Remarks"] = df.Grade.map(remarks)
df
```
